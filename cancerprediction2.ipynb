```python
data = pd.read_csv('data.csv')
display (data)
```

```python
x= data.drop ('diagnosis',axis =1).drop('id',axis =1)
display (x)
```

```python
y = data['diagnosis']
display (y)
```

```python
y = y.astype('category')
display (y)
```

```python
y = y.cat.codes
display (y)
```

```python
load_model =pickle.load(open("XGBoost","rb"))
pred1 = load_model.predict (x)
print (load_model.best_params_)
print (accuracy_score (pred1,y))
display (pred1)
```
